```python
# Physics Simulations using Python

This notebook demonstrates how fundamental physics concepts
can be translated into executable simulations using Python.

Simulations included:
1. Projectile Motion
2. Simple Harmonic Motion
3. Damped Harmonic Oscillator
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Projectile Motion

We consider a projectile launched with initial velocity `v0`
at an angle `θ` under uniform gravitational acceleration `g`.

Equations:
- x(t) = v₀ cos(θ) t
- y(t) = v₀ sin(θ) t − (1/2) g t²



```python
g = 9.8
v0 = 20
theta = 45

theta_rad = np.radians(theta)
t_flight = (2 * v0 * np.sin(theta_rad)) / g
t = np.linspace(0, t_flight, 100)

x = v0 * np.cos(theta_rad) * t
y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2

plt.plot(x, y)
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.title("Projectile Motion")
plt.grid(True)
plt.show()
```

## Simple Harmonic Motion

For a simple harmonic oscillator, the displacement is given by:

x(t) = A cos(ωt + φ)

where:
- A is amplitude
- ω is angular frequency
- φ is phase constant


```python
A = 1.0
omega = 2.0
phi = 0

t = np.linspace(0, 10, 1000)
x = A * np.cos(omega * t + phi)

plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Simple Harmonic Motion")
plt.grid(True)
plt.show()
```

## Damped Harmonic Oscillator

Real oscillatory systems lose energy over time.
The displacement is:

x(t) = A e^(−γt) cos(ωt)

where γ is the damping coefficient.


```python
A = 1.0
gamma = 0.2
omega = 2.0

t = np.linspace(0, 10, 1000)
x = A * np.exp(-gamma * t) * np.cos(omega * t)

plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Damped Harmonic Oscillator")
plt.grid(True)
plt.show()
```
